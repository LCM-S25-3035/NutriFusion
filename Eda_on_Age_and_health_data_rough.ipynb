```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For displaying all columns
pd.set_option('display.max_columns', None)
```

```python
# Upload your dataset from local (for Colab)
from google.colab import files
uploaded = files.upload()

# Read the CSV file
df = pd.read_csv("Age_and_health_data.csv")

# Preview the data
df.head()
```

```python
# Shape and basic info
print("Shape of dataset:", df.shape)
df.info()

# Summary statistics
df.describe(include='all')
```

```python
# Check for missing values
df.isnull().sum()

# Check for duplicates
df.duplicated().sum()

# Drop duplicates if needed
# df = df.drop_duplicates()
```

```python
top_n = 10
top_breakfasts = df['Breakfast Suggestion'].value_counts().nlargest(top_n).index

# Filter the dataframe
filtered_df = df[df['Breakfast Suggestion'].isin(top_breakfasts)]

plt.figure(figsize=(10, 6))
sns.countplot(y='Breakfast Suggestion', data=filtered_df, order=top_breakfasts)
plt.title(f"Top {top_n} Breakfast Suggestions")
plt.xlabel("Count")
plt.ylabel("Breakfast Suggestion")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='Breakfast Suggestion', data=df, order=df['Breakfast Suggestion'].value_counts().index)
plt.title("Distribution of Breakfast Suggestion")
plt.xlabel("Breakfast Suggestion")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Group less frequent categories into "Others"
threshold = 20  # You can tweak this
value_counts = df['Breakfast Suggestion'].value_counts()
df['Breakfast Category'] = df['Breakfast Suggestion'].apply(
    lambda x: x if value_counts[x] > threshold else 'Others'
)

plt.figure(figsize=(10, 6))
sns.countplot(y='Breakfast Category', data=df, order=df['Breakfast Category'].value_counts().index)
plt.title("Breakfast Suggestions Grouped (Rare as 'Others')")
plt.xlabel("Count")
plt.tight_layout()
plt.show()
```

```python
if 'Gender' in df.columns:
    gender_counts = df['Gender'].value_counts()
    plt.figure(figsize=(6, 6))
    plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Gender Distribution')
    plt.axis('equal')
    plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
